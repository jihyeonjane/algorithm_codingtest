## 02. 기타 기본 코드

#### time count

```python
import time

# ver01
start1 = time.time()

example1 = [i for i in range(100000) if i%2==1]
# print(example1)

end1 = time.time()
print('소요시간 ver01 :', end1 - start1)



# ver02
start2 = time.time()

example2 = []
for i in range(100000):
    if i%2==1:
        example2.append(i)
# print(example2)

end2 = time.time()
print('소요시간 ver02 :', end2 - start2)


# 시간 효율성을 위한 시간 체크 가능함....
```

#### 자료형 관련 특징
- 기본적인데 매번 헷갈림... 실제 데이터 많이 만져보는게 답

1) 순서 있음(인덱싱 가능)
- list(중복 가능) : 흔하게 많이 쓰이긴 하는데(df에도) 계산 효율성이 좋진 않은 듯
- tuple(중복 가능) : 순서 있는데 수정이 안됨. 수정이 필요 없거나, 수정하면 안되는 유형의 데이터에서 많이 쓰임

2) 순서 없음(인덱싱 불가)
- dict : key, value 개념 중요. 수정 가능함
- set(중복 불가) : 수정 가능함. 고유값을 가지므로 key 역할 잘 하는 듯함
